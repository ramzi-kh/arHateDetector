```python
def clean_text(text):
    text = re.sub(r'[.,\"،\/#!$%\^&\*;:{}=\-_`~()?؟﴾﴿]',' ',text) #remove punctuation
    text = re.sub('#\d+K\d+', ' ', text)  # years like 2K19
    text = re.sub('http\S+\s*', ' ', text)  # remove URLs
    text = re.sub('RT|cc', ' ', text)  # remove RT and cc
    text = re.sub(r'[0-9]+',' ',text) #remove numbers
    text = re.sub(r'[\u0660-\u0669]+',' ',text) #remove arabic numbers
    text = re.sub('\s+', ' ', text).strip() #remove additional whitespaces
    text = re.sub(r"_",' ',text) #remove _
   
    return text

tweet = u' جهود تبذل عاما بعد عام تعبر عن  ضىوف  تطبىق تالعام الماضى، ىنطلق الىوم لىشارك الحجاج اىامهم الروحانىه  ترجمه  الارشادىه فى منطقه  الى انترنت .  '

clean_text(tweet)
```

```python
def remove_non_arabic_letters(text):
    ENGLISH_CHARS =r'[a-zA-Z]+\b(?<!urllink|mention)'
    text = re.sub(ENGLISH_CHARS, "",  text)
    return text

tweet = u'جهود تُبذل عامًا بعد عام تعبّر عن ترابط الجسد الواحد لخدمة ضيوف الرحمن؛ تطبيق ترجمان الفائز في #هاكاثون_الحج العام الماضي،hello there ينطلق اليوم ليشارك  @reem. #حج_ذكي'


remove_non_arabic_letters(tweet)
```

```python
import re

text = u'This dog \U0001f602'
print(text)


def remove_imoji(text):

    emoji_pattern = re.compile("["
                                    u"\U0001F600-\U0001F64F"  # emoticons
                                    u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                                    u"\U0001F680-\U0001F6FF"  # transport & map symbols
                                    u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                                    u"\U00002702-\U000027B0"
                                    u"\U000024C2-\U0001F251"
                                    "]+", flags=re.UNICODE)
    text = re.sub(emoji_pattern, '', text)
    return text
   
remove_imoji(text)
```
